### Results.Plotting


#### Program for plotting results of exp3_Beforeselection- action vector analysis

```python
import matplotlib.pyplot as plt 
   
layers=range(1,10)    
adj_rand_score = [0.12,0.28,0.28,0.28,0.29,0.32,0.38,0.22,0.12] 
mut_info_score = [0.16,0.30,0.25,0.25,0.30,0.36,0.36,0.24,0.18]
homogeniety_score=[0.20,0.33,0.28,0.28,0.33,0.39,0.39,0.27,0.22]   
    
plt.rcParams.update({'font.size': 18})
plt.plot(layers, adj_rand_score, label='Adjusted Random score [-1,1]',color='r') 
plt.hlines(-0.00,1,10,color='r',linestyle='--',label='Baseline')
plt.plot(layers, mut_info_score, label='Mutual information score [-1,1]',color='b') 
plt.hlines(-0.00,1,10,color='b',linestyle='--',label='Baseline')
plt.plot(layers, homogeniety_score, label='Homogeneity score  [0,1]',color='g') 
plt.hlines(0.03,1,10,color='g',linestyle='--',label='Baseline')

plt.xlabel('Layers') 
plt.ylabel('Clustering accuracy') 
plt.legend(loc='lower right')
plt.rcParams["figure.figsize"] = (10,8)
plt.show() 
```
