```python
#### Rotating frames

import cv2
import os
from PIL import Image
import numpy as np
from skimage import io

Dir='/nobackup/leopauly/Activity_Analysis_new/Frames/Pushing/V1/'
Sav='/nobackup/leopauly/Activity_Analysis_new/Frames/Pushing/V2_from_V1/'
```

```python
count=0
files=os.listdir(Dir)
files=sorted(files)
print(files)
for i in range(len(files)):
    if (files[i].endswith('.png')):
        colorImage  = Image.open(str(Dir)+files[i])
        rot_image    = colorImage.rotate(180)
        io.imsave(str(Sav)+files[i], rot_image)
        print(files[i])
        print(files[i])
        print(count)
        count=count+1
        
```
