```python
## Imports
from keras.layers.convolutional import Convolution3D, MaxPooling3D, ZeroPadding3D
from keras.layers.core import Dense, Dropout, Flatten
from keras.models import Sequential
import random
import numpy as np
from PIL import Image
from os import listdir
from scipy.ndimage import imread
import keras
from sklearn.utils import shuffle


# Custom scripts
import lscript as lsp
import modelling as md
#from DataSet.DataSet import DataSet
import dataset as dset
import ucf101_dataset as ucf
```

```python
height=112 
width=112 
channel=3
nb_classes=101
crop_size=112

cluster_length=16
next_batch_start=0
batch_size=2
memory_batch_size_train=10
memory_batch_size_test=500
total_train_videos=13000
epochs=10
iterations=2 #total_train_videos/memory_batch_size_train
```

#### Getting the model

```python
model = md.C3D_ucf101_training_model()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

#### Training 

```python
def testing(iterations):
    test_images_neg, test_labels_uncat, next_batch_start, _, _ = ucf. read_clip_and_label('UCF101_data_preparation/test.list',memory_batch_size_test,next_batch_start,16,112,shuffle=False)
    test_images=1-test_images_neg
    test_labels=keras.utils.to_categorical(test_labels_uncat, num_classes=nb_classes)
    print(test_images.shape)
    print(test_labels_uncat)
    score = model.evaluate(test_images, test_labels, batch_size=50)
    print('testing accuracy after iteration:',iterations,score)
```

```python
next_batch_start=0
for i in range(iterations):
    train_images_neg, train_labels_uncat, next_batch_start, _, _ = ucf. read_clip_and_label('UCF101_data_preparation/train.list',memory_batch_size_train,next_batch_start,cluster_length,crop_size,shuffle=False)
    train_images=1-train_images_neg
    train_labels=keras.utils.to_categorical(train_labels_uncat, num_classes=nb_classes)
    
    print(train_images.shape)
    print(train_labels_uncat)
    for i in range(10):
        train_images, train_labels = shuffle(train_images, train_labels, random_state=i)
        
    model.fit(train_images, train_labels,batch_size=batch_size, epochs=epochs, verbose=1,validation_split=0.1, validation_data=None, shuffle=False)
    #print(next_batch_start)
    testing(iterations)
```
