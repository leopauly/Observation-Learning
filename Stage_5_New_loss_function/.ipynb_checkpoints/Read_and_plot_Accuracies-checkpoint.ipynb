```python
## Imports 
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
plt.rcParams["font.family"] = "serif"
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams["figure.figsize"] = [10, 10]
```

```python
with open("C3D_MIME20_train_90_10_shuffle_storing_accuracies.txt") as f:
    data = f.read()

data = data.split('\n')
data=data[0:len(data)-1]

print('UCF101 accuraccies', [(row) for row in data])
y = [float(row) for row in data]
x = range(0, len(y))


#plt.title("Validation accuracy during training")    
plt.xlabel('Epoches')
plt.ylabel('Accuracy (%) ')
plt.plot(x,y, c='brown', linewidth=3, label='MIME dataset')


with open("2.C3D_ucf_model_training_multigpu_py_k80_1_storing_accuracies.txt") as f:
    data = f.read()
data = data.split('\n')
data=data[0:len(data)-1]

print('UCF101 accuraccies', [(row) for row in data])
y = [float(row) for row in data]
x = range(0, len(y))

plt.plot(x,y, c='green', linewidth=3, label='UCF101 dataset')
plt.legend()

plt.show()
```
